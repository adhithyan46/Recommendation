```python
m=df['Number of google review in lakhs'].quantile(0.9)
m
```

```python
top_places = df.copy().loc[df['Number of google review in lakhs']>=m]
top_places.shape
```

```python
def weighted_rating(x,c=c,m=m):
    v=x['Number of google review in lakhs']
    r=x['Google review rating']
    return(v/(v+m)*r)+(m/(v+m)*c)
```

```python
top_places['score']=top_places.apply(weighted_rating,axis=1)
top_places.head()
```

```python
top_places=top_places.sort_values('score',ascending=False)
top_places.head()
```

```python
index = pd.Series(df.index,index=df['Name']).drop_duplicates()
index.head()
```

```python
def recommend(title,place_similarity=place_similarity):
    # title=title.lower()
    idx=index[title]
    similarity_score = list(enumerate(place_similarity[idx]))
    similarity_score=sorted(similarity_score,key=lambda x:x[1],reverse=True)
    similarity_score=similarity_score[1:11]
    place_idx=[i[0] for i in similarity_score[1:11]]
    return df['Name'].iloc[place_idx]
```

```python
recommend('India Gate')
```
