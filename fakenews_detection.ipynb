```python
import pandas as pd
import numpy as np
true_df=pd.read_csv('True.csv')
true_df.head()
```

```python
fake_df= pd.read_csv('Fake.csv')
fake_df.head()
```

```python
fake_df['label']=0
true_df['label']=1
```

```python
fake_df.head()
```

```python
fake_df.drop(['title','subject','date'],axis=1)
```

```python
true_df.drop(['title','subject','date'],axis=1)
```

```python
true_fake=pd.concat([true_df,fake_df],axis=0)
true_fake.head()
```

```python
true_fake=true_fake.sample(frac=1)
true_fake.reset_index(inplace=True)
true_fake.head()
```

```python
true_fake.drop(['index'],axis=1)
```

```python
import re
import string

def wordopt(t):  
    # Convert text to lowercase
    t = t.lower()
    # Remove text within square brackets
    t = re.sub(r'\[.*?\]', '', t)
    # Remove URLs
    t = re.sub(r'https?://\S+|www\.\S+', '', t)
    # Remove HTML tags
    t = re.sub(r'<.*?>', '', t)    
    # Remove non-alphanumeric characters (keeping spaces)
    t = re.sub(r'\W+', ' ', t)    
    # Remove newline characters
    t = re.sub(r'\n', ' ', t)    
    # Remove words containing digits
    t = re.sub(r'\b\w*\d\w*\b', '', t)
    # Remove extra spaces resulting from removal of punctuation, digits, etc.
    t = re.sub(r'\s+', ' ', t).strip()    
    return t
    
```

```python
x=true_fake['text']
y=true_fake['label']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_test.shape,y_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
x_train=tfidf.fit_transform(x_train)
x_test=tfidf.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
y_pred
```
